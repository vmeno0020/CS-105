## Getting Familiar with the Data/Data Collecting


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
NBAdata_df = pd.read_csv('final project data 1 - Sheet1.csv')
NBAdata_df
```

## Data Cleaning

First, we read in the data that we are using to understand basketball trends and changes in game play over the years. We are mainly observing the three point attempt, '3P', column to see these changes per season. We needed to remove the first two rows to display and run the dataset as it was posing issues. The first row was blank and the other was just a label; those were being counted as cells in the csv file and was showing up in the graphs. They are not needed in our dataset. 

Below, we fill the missing values with the mean values since values range similary.

```python
NBAdata_df.fillna(NBAdata_df.mean())
```

## Data Cleaning-part 2
Over here we delete row 1 for year 2021 because the year is still going and the data is incomplete for the current year.

```python
NBAdata_df = NBAdata_df.iloc[1: , :]
print("Modified Dataframe : ")
print(NBAdata_df)
```

## EDA


```python
from matplotlib.pyplot import figure

#Horizontal Bar graph (year vs 3 point attempts)
plt.bar(NBAdata_df['Season'], NBAdata_df['3PA'])
plt.xlabel('NBA Season')
plt.ylabel('3 Point Attempts')
plt.title('3 Point Attempts over the years')
plt.show()
```

The graph above shows the average 3 point attempts taken per team from 1990 onwards. As teams have taken more 3 pointers per game consistently, we can conclude a change in shot selection.

```python
#%ORB% per season%
import matplotlib.pyplot as plt

plt.plot(NBAdata_df['Season'],NBAdata_df['ORB%'])
plt.title('ORB% per Season')
plt.xlabel('Season')
plt.ylabel('ORB%')
plt.show()
```

The line graph above shows the average ratio of offensive rebounds to total rebounds per team over the years. Faster gameplay and a spread offense has the team focusing more on getting back on defense, which gives players less chance to grab offensive rebounds.

```python
#%Free Throws Ratio per season%
plt.scatter(NBAdata_df['Season'],NBAdata_df['FT/FGA'])
plt.title('Ratio of Free Throws to Field Goals')
plt.xlabel('Season')
plt.ylabel('FT/FGA')
plt.show()
```

The scatter plot above shows the average ratio of field goals to total shot attempts per team over the years. Even though the field attempts have stayed the same, a lower ratio indicates a faster and less physical gameplay. 

```python
#Line graph - offensive rating and points per game (year vs offensive rating) and (year vs points per game)

#year vs offensive rating - ORtg
years = NBAdata_df['Season']
ofs = NBAdata_df['ORtg']
df1 = pd.DataFrame({'Year': years, 'Offensive Rating': ofs})
df1.plot.line(x="Year")
```

The line graph above shows the average offensive rating per team over the years. As there is a slight positive trend, we can say that changing the shot selection has made offenses more efficient.

```python
#year vs points per game - PTS
years = NBAdata_df['Season']
pts = NBAdata_df['PTS']
df2 = pd.DataFrame({'Year': years, 'PTS': pts})
df2.plot.line(x="Year")
```

The line graph above shows the average points per game per team over the years. The trend indicates that offenses are playing at a faster pace and scoring more points due to emphasizing the 3 point shot.


```python
plt.plot(NBAdata_df['Season'], NBAdata_df['PTS']) # plot first line
plt.plot(NBAdata_df['Season'], NBAdata_df['ORtg']) # plot second line
```

This is a double line graph that is made by combining the data points from the above two line graphs to make it easier to compare.

```python
#Stacked bar graph (year vs (3pts and (total shot attempts - 3pt)
#Stacked bar graph (year vs (3pts and (total shot attempts - 3pt)
#year - x 
#shot attempts - y
#3 point attempt 
#total shot attempt - 3 pt

fga = NBAdata_df['FGA']
threegpa = NBAdata_df['3PA']
fga3pa = NBAdata_df['FGA'] - NBAdata_df['3PA']
years = NBAdata_df['Season']
df3 = pd.DataFrame({'3PA': threegpa, 'FGA - 3PA': fga3pa})
graph3 = df3.plot.bar(stacked=True, subplots=False, title='Year vs Points')
graph3.set_xlabel("Year")
graph3.set_ylabel("Points")
```

The stacked bar graph above shows the average ratio of 3 pointers to field goal attempts per team (2020 is on the left). As teams have taken a higher proportion of 3â€™s, this indicates that teams are favoring the 3 point shot.

```python
#Using linear regression, we want to predict future three points per game, 3 pt attempts, pace in the future seasons
from sklearn.linear_model import LinearRegression


tpa_train = NBAdata_df
tpa_test = NBAdata_df
tpa_train.plot.scatter(x="Season", y="3PA",fontsize = 15,figsize = (15,10))


tpa_train = tpa_train[tpa_train['3PA'].notna()]
tpa_test = tpa_train[tpa_train['3PA'].notna()]

tpa_X_train = tpa_train[["Season"]]
tpa_X_test = tpa_test[["Season"]]
tpa_y_train = tpa_train["3PA"]

lin_model = LinearRegression()
lin_model.fit(X=tpa_X_train, y=tpa_y_train)
lin_model.predict(X=tpa_X_test)
```

```python
#Part with the actual line
tpa_X_new = pd.DataFrame()
# create a sequence of 200 evenly spaced numbers from 10 to 41
tpa_X_new["Season"] = np.linspace(1990, 2027, num=100)

# create a Series out of the predicted values
# (trailing underscore indicates fitted values)
tpa_y_new_ = pd.Series(
    lin_model.predict(tpa_X_new), # y values in Series.plot.line()
    index=tpa_X_new["Season"]    # x values in Series.plot.line()
)

# plot the data, then the model
tpa_train.plot.scatter(x="Season", y="3PA", figsize = (30,60),fontsize = 20)
tpa_y_new_.plot.line(figsize = (30,10),fontsize = 20, color="red")
```

From the linear analysis we see that there is a clear upward trend in 'Three point attempts' throughout the seasons and has been growing over the past few years.

```python
#Using linear regression 3P in the future seasons
from sklearn.linear_model import LinearRegression

tpa_train = NBAdata_df
tpa_test = NBAdata_df
tpa_train.plot.scatter(x="Season", y="3P",fontsize = 20,figsize = (25,10))


tpa_train = tpa_train[tpa_train['3P'].notna()]
tpa_test = tpa_train[tpa_train['3P'].notna()]

tpa_X_train = tpa_train[["Season"]]
tpa_X_test = tpa_test[["Season"]]
tpa_y_train = tpa_train["3P"]

lin_model = LinearRegression()
lin_model.fit(X=tpa_X_train, y=tpa_y_train)
lin_model.predict(X=tpa_X_test)
```

```python
#Part with the actual line
tpa_X_new = pd.DataFrame()
# create a sequence of 200 evenly spaced numbers from 10 to 41
tpa_X_new["Season"] = np.linspace(1990, 2030, num=100)

# create a Series out of the predicted values
# (trailing underscore indicates fitted values)
tpa_y_new_ = pd.Series(
    lin_model.predict(tpa_X_new), # y values in Series.plot.line()
    index=tpa_X_new["Season"]    # x values in Series.plot.line()
)

# plot the data, then the model
tpa_train.plot.scatter(x="Season", y="3P", figsize = (30,60),fontsize = 20)
tpa_y_new_.plot.line(figsize = (25,10),fontsize = 20, color="red")
```

The graph above shows linear regression of the 3 point shots scored over the course of the season going up to 2027. Following the line, we can see that the more 3 point shots will continue to be made.
