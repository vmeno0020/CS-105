```python
#Using linear regression, we want to predict future three points per game, 3 pt attempts, pace in the future seasons
from sklearn.linear_model import LinearRegression


tpa_train = NBAdata_df
tpa_test = NBAdata_df
tpa_train.plot.scatter(x="Season", y="3PA",fontsize = 15,figsize = (15,10))


tpa_train = t_train[t_train['3PA'].notna()]
tpa_test = t_train[t_train['3PA'].notna()]

tpa_X_train = tpa_train[["Season"]]
tpa_X_test = tpa_test[["Season"]]
tpa_y_train = tpa_train["3PA"]

lin_model = LinearRegression()
lin_model.fit(X=tpa_X_train, y=tpa_y_train)
lin_model.predict(X=tpa_X_test)
```

```python
#Part with the actual line
tpa_X_new = pd.DataFrame()
# create a sequence of 100 evenly spaced years from 1990 to 2027
tpa_X_new["Season"] = np.linspace(1990, 2027, num=100)

# create a Series out of the predicted values
# (trailing underscore indicates fitted values)
tpa_y_new_ = pd.Series(
    lin_model.predict(tpa_X_new), # y values in Series.plot.line()
    index=tpa_X_new["Season"]    # x values in Series.plot.line()
)

# plot the data, then the model
tpa_train.plot.scatter(x="Season", y="3PA", figsize = (30,60),fontsize = 20)
tpa_y_new_.plot.line(figsize = (30,10),fontsize = 20, color="red")
```
